```python
import glob
folder_path = "/Users/apple/Desktop/deep_eye/dataset/raw_data/"
#folder_content = sorted(os.listdir(folder_path)) # all file in folder
folder_content = sorted(glob.glob(folder_path + "/*.csv")) # all file in folder

QPC_AOI_df = pd.read_csv("/Users/apple/Desktop/deep_eye/dataset/AOI/origin/P1_Q_PIC_OPTION_AOI.csv").iloc[:, 1:]

totalpq_QCP_sequence = []
totalpq_QCP_sequence_df = pd.DataFrame()
for p in  range(len(folder_content)): # p=60
    df = pd.read_csv(folder_content[p])
    df_col = df[['CURRENT_FIX_START','CURRENT_FIX_X','CURRENT_FIX_Y','CURRENT_FIX_DURATION']]
    df_col.columns = ['start_time', 'x', 'y', 'duration']

    totalq_QCP_sequence = []
    totalq_QCP_sequence_df = pd.DataFrame()
    for q in range(47): # for q in range(1,48):
        if int(resp_array[p][q]) == 0:
            df_time = df_col[(df_col['start_time']>= int(onset_array[p][q])) & (df_col['start_time']<= int(onset_array[p][q])+14000)] # Q1,Q2...
        else:
            df_time = df_col[(df_col['start_time']>= int(onset_array[p][q])) & (df_col['start_time']<= int(resp_array[p][q]))] # Q1,Q2...
        df_time = df_time[(df_time['duration']>= 100)] # duration >= 100

        # x and y QCP coordinate    
        x_cor = QPC_AOI_df.iloc[:,[0,2,4,6,8,10]]
        y_cor = QPC_AOI_df.iloc[:,[1,3,5,7,9,11]]
    
        Q = df_time[(df_time['x'] >= int(x_cor.iloc[q][0])) & (df_time['x'] <= int(x_cor.iloc[q][1])) & (df_time['y'] >= int(y_cor.iloc[q][0])) & (df_time['y'] <= int(y_cor.iloc[q][1]))]
        C = df_time[(df_time['x'] >= int(x_cor.iloc[q][2])) & (df_time['x'] <= int(x_cor.iloc[q][3])) & (df_time['y'] >= int(y_cor.iloc[q][2])) & (df_time['y'] <= int(y_cor.iloc[q][3]))]
        P = df_time[(df_time['x'] >= int(x_cor.iloc[q][4])) & (df_time['x'] <= int(x_cor.iloc[q][5])) & (df_time['y'] >= int(y_cor.iloc[q][4])) & (df_time['y'] <= int(y_cor.iloc[q][5]))]

        Q = Q.assign(label = ["Q"]*Q.shape[0])
        C = C.assign(label = ["C"]*C.shape[0])
        P = P.assign(label = ["P"]*P.shape[0])
        QCP_df = pd.concat([Q, C, P], axis = 0)
        QCP_df = QCP_df.sort_values('start_time') # QCP_df in time order 
        
        QCP_sequence = []
        for i in range(QCP_df.shape[0]):
            repeat_num = int(np.floor(QCP_df.iloc[i][3]/100))
            QCP_sequence.append(QCP_df.iloc[i][4]*repeat_num)
        totalq_QCP_sequence.append(QCP_sequence) # for 60p*47q df
        
    totalpq_QCP_sequence.append(totalq_QCP_sequence) # for 60p*47q df
    print("ID", p+1 ,":",len(totalq_QCP_sequence))

totalpq_QCP_sequence_tempform = pd.DataFrame(totalpq_QCP_sequence) # 60p*47q df
totalpq_QCP_sequence_tempform.to_csv("/Users/apple/Desktop/deep_eye/dataset/QCP_sequence/AC_totalpq_QCP_sequence_tempform.csv")# 60p*47q df
```

```python
import tensorflow as tf
from tf.keras.datasets import imdb
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers.embeddings import Embedding
from keras.preprocessing import sequence
# fix random seed for reproducibility
numpy.random.seed(7)
```

```python
# load the dataset but only keep the top n words, zero the rest
top_words = 5000
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=top_words)
```

```python
print()
```

```python
# truncate and pad input sequences
max_review_length = 500
X_train = sequence.pad_sequences(X_train, maxlen=max_review_length)
X_test = sequence.pad_sequences(X_test, maxlen=max_review_length)
```

```python
# create the model
embedding_vecor_length = 32
model = Sequential()
model.add(Embedding(top_words, embedding_vecor_length, input_length=max_review_length))
model.add(LSTM(100))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
print(model.summary())
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=3, batch_size=64)
```

```python
# Final evaluation of the model
scores = model.evaluate(X_test, y_test, verbose=0)
print("Accuracy: %.2f%%" % (scores[1]*100))
```

```python
# LSTM for sequence classification in the IMDB dataset
import numpy
from keras.datasets import imdb
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers.embeddings import Embedding
from keras.preprocessing import sequence
# fix random seed for reproducibility
numpy.random.seed(7)
# load the dataset but only keep the top n words, zero the rest
top_words = 5000
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=top_words)
# truncate and pad input sequences
max_review_length = 500
X_train = sequence.pad_sequences(X_train, maxlen=max_review_length)
X_test = sequence.pad_sequences(X_test, maxlen=max_review_length)
# create the model
embedding_vecor_length = 32
model = Sequential()
model.add(Embedding(top_words, embedding_vecor_length, input_length=max_review_length))
model.add(LSTM(100))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
print(model.summary())
model.fit(X_train, y_train, epochs=3, batch_size=64)
# Final evaluation of the model
scores = model.evaluate(X_test, y_test, verbose=0)
print("Accuracy: %.2f%%" % (scores[1]*100))
```
